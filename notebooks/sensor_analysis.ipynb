# Sensor Anomaly Analysis
Generate synthetic data, fit hybrid detector, visualize results.

```python
import sys; sys.path.append('..')
from src.generate_signals import generate
from src.hybrid_detector import HybridDetector
df = generate(3000)
df.head()
```

```python
det = HybridDetector().fit(df)
scored = det.score(df)
print('Anomaly rate:', scored['is_anomaly'].mean())
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12, 4))
ax.scatter(scored['ts'], scored['voltage'], c=scored['is_anomaly'].map({True:'red', False:'gray'}), s=6, alpha=0.6)
ax.set_title('Voltage with anomalies highlighted')
plt.show()
```

```python
import pandas as pd
pd.crosstab(scored['is_truth_anomaly'], scored['is_anomaly'])
```
